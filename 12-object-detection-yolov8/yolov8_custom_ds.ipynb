```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

```python
model = YOLO('yolov8n.pt')
```

```python
model.train(data='dataset/data.yaml', epochs=50, imgsz=640, batch=16, project='results', device='cpu')
```

```python

model = YOLO('results/train/weights/best.pt')
results = model.predict(source='data/val/images/', imgsz=640, conf=0.25, device='cpu')
```

```python
# Loop through each result (each image/frame)
for i, result in enumerate(results):
    # result.plot() returns an image with boxes drawn (as numpy array in BGR format)
    img_with_boxes = result.plot()

    # Convert BGR (OpenCV format) to RGB for matplotlib display
    img_rgb = cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB)

    # Show image with bounding boxes
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.title(f'Detection Result {i + 1}')
    plt.show()

    # Optionally save the image with boxes
    cv2.imwrite(f'result_{i+1}.jpg', img_with_boxes)
```
